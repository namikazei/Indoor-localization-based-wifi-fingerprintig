# Project Background & Data

Indoor locationing is one of the many contemporary research topics that is being investigated since GPS signals are generally too week inside large industrial complexes. There are numerous applications for indoor locationing, but the one that interests me the most is navigational aide to those who are visually impaired or blind. The technique applied in this project uses Wi-Fi signal from a network router to a smart device, such as a smart cell phone or a tablet.  

The data for this project is taken from [UCI Machine Learning Repository's UJIIndoorLoc Data Set](http://archive.ics.uci.edu/ml/datasets/UJIIndoorLoc).  The data were collected at the three multi-floor buildings of the Jaume I University (UJI). The main characteristics of the data are:

* Surface area = 108,703 $m^2$
* Buildings = 3
* Floors = maximum 5
* The number of different places (reference points) = 933
* Samples = 21,049 (19,938 for training/learning and 1,111 for testing/validation)
* Testing samples collected four months after the training samples
* Wireless access points (WAPs) = 520
* Data were collected by more than 20 users using 25 different model of mobile devicesd (some users used more than one model).

Description of elements in the data:
* 001 - 520: RSSI levels
> 520-element vector from each record contains the raw intensity levels of the detected WAP from a single WiFi scan. RSSI levels correspond to negative integer values measured in dBm, where -100dBm is equivalent to a very weak signal, whereas 0dBm means that the detected WAP has an extremely good signal. Not all WAPs are detected in each scan. WAPs that have not been detected by the device are artificially labeled +100dBm.
* 521 - 522: Real world coordinates of the sample points
>Longitude and latitude coordinates.
* 523: Floor
>Floor number in the building.
* 524: Building ID
>Corresponds to the building in which the capture was taken.
* 525: Space ID
> Identifies space (offices, labs, classroom, etc.) where the capture was taken.
* 526: Relative position with respect to Space ID
> Binary values indicating positive outside or inside the Space ID.
* 527: User ID
> Represents 18 different users.
* 528: Phone ID
>Represents the Android device used.
* 529: Timestamp
> Time (in Unix time format) when the capture was taken.

# Preprocess

###### Copy of training data set

```python
# Create a data set to preprocess by creating a copy with indicies (deep=True)

trainingDF = training.copy(deep=True)
```

###### Filter data to create data set for each building 

```python
# Filter to only include building id = 0
# Model only needs to train inside a single building, 
# GPS can be used to navigate from building to building

trainingDF_B0 = trainingDF[trainingDF['BUILDINGID'] == 0]
trainingDF_B1 = trainingDF[trainingDF['BUILDINGID'] == 1]
trainingDF_B2 = trainingDF[trainingDF['BUILDINGID'] == 2]
```

###### Create a unique dependent position feature for each building data set

```python
# Create a unique id of position
# Unique position is a combination of floor, relativeposition, and spaceid

trainingDF_B0['u_id'] = trainingDF_B0['FLOOR'].map(str) + trainingDF_B0['SPACEID'].map(str) + trainingDF_B0['RELATIVEPOSITION'].map(str)
trainingDF_B1['u_id'] = trainingDF_B1['FLOOR'].map(str) + trainingDF_B1['SPACEID'].map(str) + trainingDF_B1['RELATIVEPOSITION'].map(str)
trainingDF_B2['u_id'] = trainingDF_B2['FLOOR'].map(str) + trainingDF_B2['SPACEID'].map(str) + trainingDF_B2['RELATIVEPOSITION'].map(str)
```

```python
trainingDF_B0
```

###### Remove unnecessary features

```python
# Remove features that are not part of the position
# Longitude, latitude, floor, buildingid, spaceid, relativeposition,, userid, phoneid, timestamp

trainingDF_B0.drop(['LONGITUDE', 'LATITUDE', 'FLOOR', 'BUILDINGID', 'SPACEID', 'RELATIVEPOSITION', 'USERID', 'PHONEID', 'TIMESTAMP'], axis=1, inplace=True)
trainingDF_B1.drop(['LONGITUDE', 'LATITUDE', 'FLOOR', 'BUILDINGID', 'SPACEID', 'RELATIVEPOSITION', 'USERID', 'PHONEID', 'TIMESTAMP'], axis=1, inplace=True)
trainingDF_B2.drop(['LONGITUDE', 'LATITUDE', 'FLOOR', 'BUILDINGID', 'SPACEID', 'RELATIVEPOSITION', 'USERID', 'PHONEID', 'TIMESTAMP'], axis=1, inplace=True)
```

###### Convert the dependent feature to categorical

```python
trainingDF_B0['u_id'] = pd.Categorical(trainingDF_B0['u_id'])
trainingDF_B1['u_id'] = pd.Categorical(trainingDF_B1['u_id'])
trainingDF_B2['u_id'] = pd.Categorical(trainingDF_B2['u_id'])
```

```python
trainingDF_B0
```

###### Save and load the processed data set

```python
# Save the processed file

trainingDF_B0.to_csv('trainingDF_B0.csv')
trainingDF_B1.to_csv('trainingDF_B1.csv')
trainingDF_B2.to_csv('trainingDF_B2.csv')
```

```python
# Load the saved data set

trainingDF_B0 = pd.read_csv('trainingDF_B0.csv', header = 0)
trainingDF_B1 = pd.read_csv('trainingDF_B1.csv', header = 0)
trainingDF_B2 = pd.read_csv('trainingDF_B2.csv', header = 0)
```

# Model Development

###### Create independent/dependent variable

```python
# Ind/Dep variables for Building 0
X = trainingDF_B0.iloc[:, 0:520]
y = trainingDF_B0.iloc[:, -1]

# Ind/Dep variables for Building 1
X_B1 = trainingDF_B1.iloc[:, 0:520]
y_B1 = trainingDF_B1.iloc[:, -1]

# Ind/Dep variables for Building 2
X_B2 = trainingDF_B2.iloc[:, 0:520]
y_B2 = trainingDF_B2.iloc[:, -1]
```

###### Split training data set 80/20 train/test

```python
# Building 0
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Building 1
X_train_B1, X_test_B1, y_train_B1, y_test_B1 = train_test_split(X_B1, y_B1, test_size=0.2, random_state=0)

# Building 2
X_train_B2, X_test_B2, y_train_B2, y_test_B2 = train_test_split(X_B2, y_B2, test_size=0.2, random_state=0)
```
